Los diferentes tipos de uniones en pandas son los siguientes: 
Pandas ofrece varios métodos para unir y combinar tablas (DataFrames). Los más comunes son merge, join y concat. Cada uno tiene sus propias características y usos específicos. A continuación, se describen estos métodos y sus diferencias, junto con ejemplos de sintaxis.

### 1. merge
El método merge es similar a las uniones en SQL. Permite unir dos DataFrames basándose en una o más claves.

Inner join (intersección): Devuelve solo las filas que tienen valores coincidentes en ambas tablas.
Outer join (unión): Devuelve todas las filas cuando hay una coincidencia en una de las tablas.
Left join: Devuelve todas las filas del DataFrame izquierdo y las filas coincidentes del DataFrame derecho.
Right join: Devuelve todas las filas del DataFrame derecho y las filas coincidentes del DataFrame izquierdo.
### 2. join
El método join se utiliza principalmente para combinar DataFrames por sus índices. Puede ser más conveniente cuando se trabaja con índices en lugar de columnas clave.

Inner join (intersección)
Outer join (unión)

### 3. concat

El método concat se utiliza para concatenar DataFrames a lo largo de un eje (filas o columnas).
Resumen de diferencias:
merge: Similar a las uniones en SQL, combina DataFrames basándose en claves específicas. Se utiliza cuando se necesita combinar DataFrames basados en columnas.
join: Similar a merge, pero se usa principalmente para unir DataFrames basándose en índices. Conveniente cuando se trabaja con índices.
concat: Utilizado para concatenar DataFrames a lo largo de un eje (filas o columnas). Es útil para agregar más datos a un DataFrame existente.
Cada uno de estos métodos ofrece flexibilidad para diferentes tipos de combinaciones y es importante elegir el adecuado según el caso de uso específico.

## Justificación
